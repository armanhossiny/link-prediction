```python
import matplotlib.pyplot as plt
import networkx as nx

from networkx.algorithms.community import greedy_modularity_communities
from networkx.algorithms.community.label_propagation import asyn_lpa_communities
from networkx.algorithms.community.quality import coverage
from networkx.algorithms.community.quality import modularity
from networkx.algorithms.community.quality import performance
```

```python
G =nx.Graph()

# # dataset lesmis
# //////////////////////////////////////////////////////////////////////////////////////dataset  
# edges = nx.read_gml('dataset/lesmis.gml')
# G.add_edges_from(edges.edges(data=True))

# //////////////////////////////////////////////////////////////////////////////////////dataset King_James
# G = nx.read_edgelist('dataset/King_James.txt', nodetype=int, data=(('weight',float),))
# G.edges(data=True)

# //////////////////////////////////////////////////////////////////////////////////////dataset NetScience
# G = nx.read_edgelist('dataset/NetScience.txt', nodetype=int, data=(('weight',float),))
# G.edges(data=True)


# //////////////////////////////////////////////////////////////////////////////////////dataset Adolescent
G = nx.read_edgelist('dataset/Adolescent.txt', nodetype=int, data=(('weight',float),))
G.edges(data=True)
```

```python
G.remove_edges_from(nx.selfloop_edges(G))

arrayGraph=nx.to_numpy_matrix(G,sorted(G.nodes()))
len(arrayGraph)
G = nx.from_numpy_matrix(arrayGraph)

main_G = G.copy()
main_matrix = arrayGraph
```

```python
def percentage(percent, whole):
    return int((percent * whole) / 100.0)
```

```python
def reverse_weight_edges(G_weight):
    G=[]
    min=False
    max=0
    for u, v, data in G_weight:
        if min==False:
            min=data['weight']
        if data['weight']<min:
            min=data['weight']
        if data['weight']>max:
            max=data['weight']
    for u, v, data in G_weight:
        G.append((u,v,{'weight': round((max+min)-data['weight'], 2)}))
    print("sum Min and Max weight:",max+min)
    return G
```

```python
def common_neighbors(G, u, v):
    if u not in G:
        raise nx.NetworkXError('u is not in the graph.')
    if v not in G:
        raise nx.NetworkXError('v is not in the graph.')
    CN=[]
    CN_edges=[]
    for w in G[u]:
        if w in G[v] and w not in (u,v):
            CN.append(w)
            CN_edges.append((w, u))
            CN_edges.append((w, v))

    aa = (w for w in G[u] if w in G[v] and w not in (u, v))
#     return CN, CN_edges
    return CN_edges
#     return len(CN)
```

```python
def weight_common_neighbors(G_weight, CN_edges):
        WCN=0
#     while index<=(len(CN_edges)-1):
        for u, v, data in G_weight:
            if (u, v) in CN_edges or (v, u) in CN_edges:
                WCN+=data['weight']
#                 print(u,v, ": ",data['weight'])
        return (round(WCN,2))
```

# Main Weighted Edges Graph

```python
G_weight_main=list(G.edges(data=True))
```

# Reverse Weighted Edges Graph

```python
# reverse weight edges graph
G_weight_reverse= reverse_weight_edges(G_weight_main)
```

# positive link prediction

```python
G_weight_main=list(G.edges(data=True))


CN_value=[]

Matrix=nx.to_numpy_matrix(G,sorted(G.nodes()))
maxIndex= len(G.nodes())
TopRank=-1
TempRank=0
i=0
j=0
for i in list(range((maxIndex-1))):
    for j in list(range(i,(maxIndex))):
#         print(i,j)
        if Matrix[i,j] ==0:
            TempRank = len(list(nx.common_neighbors(G, i, j)))
#             print(i,j, ": ", TempRank)
            edges= common_neighbors(G, i, j)
            CN_value.append((i,j,weight_common_neighbors(G_weight_main, edges)))
#             print("node(", i,",",j,")","number_common_neighbors:",TempRank,"weight: ", weight_common_neighbors(G_weight_main, edges))
            if TempRank> TopRank:
#                 print("TopRank: ",i, j,": ", TempRank)
                TopRank=TempRank

positive_common_neighbors_results= sorted(CN_value, reverse=True, key=lambda x: x[2])

PCN=[]
for edge in positive_common_neighbors_results:
    if edge[2]!=0:
        PCN.append((edge[0], edge[1], edge[2]))
positive_common_neighbors_results=PCN
len(positive_common_neighbors_results)
```

# Added Link to Graph

```python
print(len(G.edges()))
percentage_10 = percentage(5, len(positive_common_neighbors_results))
print("percentage: ", percentage_10)
index=1
for edge in positive_common_neighbors_results:
#     print(edge[0],edge[1], weight)
    G.add_edge(edge[0],edge[1], weight=edge[2])
    if index>=percentage_10:
        break;
    index +=1
print(len(G.edges()))
```

# negative link prediction

```python
G_weight_main=list(G.edges(data=True))

G_weight_reverse= reverse_weight_edges(G_weight_main)

CN_value=[]

Matrix=nx.to_numpy_matrix(G,sorted(G.nodes()))
maxIndex= len(G.nodes())
TopRank=-1
TempRank=0
i=0
j=0
for i in list(range((maxIndex-1))):
    for j in list(range(i,(maxIndex))):
#         print(i,j)
        if Matrix[i,j] !=0:
            TempRank = len(list(nx.common_neighbors(G, i, j)))
#             print(i,j, ": ", TempRank)
            edges= common_neighbors(G, i, j)
            CN_value.append((i,j,weight_common_neighbors(G_weight_reverse, edges)))
#             print("node(", i,",",j,")","number_common_neighbors:",TempRank,"weight: ", weight_common_neighbors(G_weight_main, edges))
            if TempRank> TopRank:
#                 print("TopRank: ",i, j,": ", TempRank)
                TopRank=TempRank

common_neighbors_results= sorted(CN_value, reverse=True, key=lambda x: x[2])
```

# Remove anomaly edges

```python
print(len(G.edges()))
def percentage(percent, whole):
    return int((percent * whole) / 100.0)

percentage_negative_10 = percentage(10, len(common_neighbors_results))
print("numbers remove edges: ", percentage_negative_10)
index=1

for edge in common_neighbors_results:
    if G.has_edge(edge[0], edge[1]):
        G.remove_edge(edge[0], edge[1])
        if index>=percentage_negative_10:
            print(index, " : ", percentage_negative_10)
            break;
        index +=1
        


print(len(G.edges()))
```
